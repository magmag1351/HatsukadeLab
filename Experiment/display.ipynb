# 実験測定結果用ソースコード

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import glob

def DisplayMeasurementResults(csvFileName1, ylim, DefectFlag=False, csvFileName2=None):

  input_filepath01 = pd.read_csv(csvFileName1, encoding='cp932')
  # input_filepath02 = pd.read_csv(csvFileName2, encoding='cp932')

  plt.figure(figsize=(14, 8))
  plt.plot(input_filepath01['時間[μs]'], input_filepath01[' 加算平均値[V]'], label='振幅', color='black')

  x_lines = [94, 169, 584, 660]
  for x in x_lines:
    plt.axvline(x=x, color='red', linestyle='--', linewidth=1) 
    plt.text(x, -ylim-0.002, f'{x}μs', color='red', fontsize=10, ha='center')

  if DefectFlag:
    x_lines_defect = [320, 396]
    for x in x_lines_defect:
      plt.axvline(x=x, color='blue', linestyle='--', linewidth=1)
      plt.text(x, -ylim-0.002, f'{x}μs', color='blue', fontsize=10, ha='center')

  plt.grid(True)
  plt.xlabel('Time [μs]')
  plt.ylabel('Amplitude [V]')
  plt.xlim(0, 700)
  plt.ylim(-ylim, ylim)
  plt.savefig('./output.png')
  plt.show()
  plt.close()
```

```python
file_path = './source/csv/比較/非貫通/単体250差動.csv'
DisplayMeasurementResults(file_path, 0.02, True)
```
